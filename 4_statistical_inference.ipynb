```python
import numpy as np
import pandas as pd
from functions import *
```

```python
df = pd.read_csv("data/gielda.csv", sep=";")
df["IndexReturnPerc"] = df["IndexReturnPerc"].apply(
    lambda x: float(x.replace(",", "."))
)
df["StockReturnPerc"] = df["StockReturnPerc"].apply(
    lambda x: float(x.replace(",", "."))
)

df.head()
```

```python
# correlation between the height and head size in the sample:
from scipy import stats

correlation_test = stats.pearsonr(df["IndexReturnPerc"], df["StockReturnPerc"])
print(correlation_test)
```

as alpha =0.1, we can reject the null hypothesis. Thre is a linear correlation between index and stock returns.

```python
confidence_interval = correlation_test.confidence_interval(0.90)

print(
    "Test:\n\n", correlation_test, "\n\n Confidence interval: \n\n", confidence_interval
)
```

we are 90% confident that correlation between index and stock returns is between 4.76% and 59.2%.
