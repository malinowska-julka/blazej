```python
import numpy as np
import pandas as pd
from functions import *
```

```python
df = pd.read_csv("data/returns.csv")
df.head()
```

```python
# Standard deviation
df["Return"].std()
```

Std interpret: data is quite spreaded, most of values are within the range of +-27 from the mean value.

```python
# Give the first quartile:

print("Q1: " + str(df["Return"].quantile(0.25)))

# Give the empirical value of VaR(0.05), i.e., the fifth percentile of returns:
var = df["Return"].quantile(0.05)
print("VaR 0.05: " + str(var))

# Give the empirical value of ES(0.05), i.e., the average return for observations below the fifth percentile
es = df["Return"].loc[df["Return"] < var].mean()
print("ES 0.05: " + str(es))
```

Q1: 25% of returns are lower than -6.25 p.p.
Value at Risk: there is 5% chance that i can have lower returns than 29.5 p.p.
Expected shortfall: during the worst 5% months (when returns < -29.5), the expected average loss is -35.83 p.p.

```python
# mean and test for the mean
mean = np.mean(df["Return"])
mean
```

```python
import scipy.stats as stats

test_result = stats.ttest_1samp(df["Return"], popmean=9, alternative="greater")

print(test_result)
```
