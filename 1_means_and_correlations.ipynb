```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/results.csv")
```

```python
print("Mean A: " + str(np.mean(df["A_error"])))
print("Mean B: " + str(np.mean(df["B_error"])))
```

```python
# Pearson correlation
df[["A_error", "B_error"]].corr()
```

```python
# Spearman correlation

df[["A_error", "B_error"]].corr(method="spearman")
```

Means: Mr A. makes guesses almost twice closer to the true age than Mrs B. On average A overestimates age by 1.55 year 
and B by 3 years.

Correlations: there is high correlation between A and B's errors, which indicates their similar guesses.
