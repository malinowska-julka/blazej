```python
import numpy as np
import pandas as pd
from functions import *
```

```python
df = pd.read_csv("data/q_data.csv", sep=";")
df.head()
```

```python
cont_table = pd.crosstab(df["Movie genre"], df["Area of study"])

# Compute chi-squared statistic
chi2 = chi2_contingency(cont_table)
print(chi2)
# Get sample size and minimum dimension
n = cont_table.sum().sum()
k = min(cont_table.shape) - 1

# Compute Cram√©r's V
v = np.sqrt(chi2[0] / (n * k))
print("Cramer's V: " + str(v))
```

Interpretation<br>
CHI2 : we cannot reject the null hypothesis, i.e. there is no difference between students from both groups in terms of preferable movie genre.
Cramers V (range from 0 to 1): 0.08 suggests very weak association between student group and movie type.
